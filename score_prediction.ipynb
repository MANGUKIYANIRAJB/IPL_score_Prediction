## 1) remove the unwanted columns

```python
df.drop(columns=['mid','venue','batsman','bowler','striker','non-striker'],axis=1,inplace=True)
```

## 2) Keep only consistance teams

```python
consistence_team= ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals','Mumbai Indians','Kings XI Punjab',
                   'Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
df=df[(df['bat_team'].isin(consistence_team))&(df['bowl_team'].isin(consistence_team))]
```

```python
df['bat_team'].unique()
```

```python
df.head()
```

## 3) Removing first 5 over Data from all matches

```python
df=df[df['overs']>=5.0]
```

```python
df.head()
```

## 4) convert string into Datetime

```python
from datetime import datetime

df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
df.head()
```

## Data Preprocessing

Handling categorical features<br/> Splitting dataset into train and test set on the basis of daa  </br>e

```python
# we will convert cetegorical feature using OnehotEncoder
encoding_df=pd.get_dummies(data=df,columns=['bat_team','bowl_team'])
```

```python
encoding_df.columns
```

```python
encoding_df.head()
```

```python
# Rearranging the columns
encoding_df = encoding_df[['date','bat_team_Royal Challengers Bangalore', 'bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils', 'bat_team_Kings XI Punjab',
              'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
              'bat_team_Sunrisers Hyderabad','bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',
              'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
              'bowl_team_Royal Challengers Bangalore', 'bowl_team_Sunrisers Hyderabad',
              'overs', 'runs', 'wickets', 'runs_last_5', 'wickets_last_5', 'total']]
```

```python
encoding_df.head()
```

## Splitting the Dataset 

â€¢ Model *trained on* the data from **IPL Seasons 1 to 9** ie: (2008 to 2016)<br/>
â€¢ Model *tested on* data from **IPL Season 10** ie: (2017)<br/>
â€¢ Model *predicts on* data from **IPL Seasons 11 to 12** ie: (2018 to 2019)

```python
x_train=encoding_df.drop(labels='total',axis=1)[encoding_df['date'].dt.year<=2016]
x_test=encoding_df.drop(labels='total',axis=1)[encoding_df['date'].dt.year>=2017]

y_train=encoding_df[encoding_df['date'].dt.year<=2016]['total'].values
y_test=encoding_df[encoding_df['date'].dt.year>=2017]['total'].values

x_train.drop(labels='date',axis=1,inplace=True)
x_test.drop(labels='date',axis=1,inplace=True)
```

## Model Selection

**Liner Regression**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
model_1_result=lr.score(x_test,y_test)
print(" an Accuracy Score = ",model_1_result)
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
model_2_result=dt.score(x_test,y_test)
print(" an Accuracy Score = ",model_2_result)
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
model_3_result=rf.score(x_test,y_test)
print(" an Accuracy Score = ",model_3_result)
```

**SVM**

```python
from sklearn.svm import SVC
```

```python
S=SVC()
```

```python
S.fit(x_train,y_train)
```

```python
model_4_result=S.score(x_test,y_test)
print(" an Accuracy Score = ",model_4_result)
```

**Logistic Regresion**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lor=LogisticRegression()
```

```python
lor.fit(x_train,y_train)
```

```python
model_5_result=lor.score(x_test,y_test)
print(" an Accuracy Score = ",model_5_result)
```

## Check the Accuracy of the Model 

```python
from sklearn.metrics import accuracy_score,r2_score,mean_absolute_error,mean_squared_error
```

```python
y1=lr.predict(x_test)
mse=mean_squared_error(y_test,y1)
mae=mean_absolute_error(y_test,y1)
rmse=np.sqrt(mse)
print('an mean absolure error is :',mae)
print('an mean squared error is :',mse)
print(" An Root mean squared error :",rmse)
```

```python
y2=dt.predict(x_test)
ya2=accuracy_score(y_test,y2)
print(" An accuracy of this model using Decision Tree :",ya2)
```

```python
y3=rf.predict(x_test)
mse=mean_squared_error(y_test,y3)
mae=mean_absolute_error(y_test,y3)
rmse=np.sqrt(mse)
r_2=r2_score(y_test,y3)
print('an mean absolure error is :',mae)
print('an mean squared error is :',mse)
print(" An Root mean squared error :",rmse)
print("an R2_score is: ",r_2)
```
